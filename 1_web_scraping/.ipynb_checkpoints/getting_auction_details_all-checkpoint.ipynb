```python
from bs4 import BeautifulSoup
import requests
import time
import re
import numpy as np
```

## loading auction list

```python
auction_list = []
```

```python
with open('apartments_links.txt','r') as file:
    for line in file:
        auction_list.append(line[:-1])
```

## collecting auction details

```python
data = []
```

```python
start = time.time()

i = 0

for url in auction_list[:1000]:
    try:
        page = requests.get(url)
        soup = BeautifulSoup(page.text, 'html.parser')
    except:
        continue
    # location
    try:
        location = soup.find_all('a',class_="parameters__locationLink")
        city = location[0].text
    except:
        city = ''
    try:
        district = location[1].text
    except:
        district = ''
    try:
        voivodeship = location[2].text
    except:
        voivodeship = ''
    # coordinates
    try:
        scripts = soup.find(id='leftColumn').find_all('script')
        localization_y = re.findall(r'(\"\w+szerokosc.*?\"):(\d*[.\d]*)', scripts[6].contents[0])[0][1]
    except:
        localization_y = ''
    try:
        localization_x = re.findall(r'(\"\w+dlugosc.*?\"):(\d*[.\d]*)', scripts[6].contents[0])[0][1]
    except:
        localization_x = ''
    # parameters
    try:
        scripts2 = soup.find('body').find_all('script')
        market = re.findall(r'(\"rynek\"):\"(.*?)\"', scripts2[-7].contents[0])[0][1]
    except:
        market = ''
    try:
        parameters = soup.find_all('ul',class_="parameters__rolled")
        area = re.findall(r'(Powierzchnia).*\n(.*)m', parameters[0].text)
        area = area[0][1].replace(' ','')
        area = area.replace(',','.')
    except:
        area = ''
    try:
        rent = re.findall(r'(Opłaty).*\n(.*)zł', parameters[0].text)
        rent = rent[0][1].replace(' ','')
        rent = rent.replace(',','.')
    except:
        rent = ''
    try:
        rooms = re.findall(r'(Liczba pokoi).*\n(.*)', parameters[0].text)
        rooms = rooms[0][1]
    except:
        rooms = ''
    try:
        floor = re.findall(r'(Piętro).*\n(.*)', parameters[0].text)
        floor = floor[0][1]
    except:
        floor = ''
    try:
        floors = re.findall(r'(Liczba pięter w budynku).*\n(.*)', parameters[0].text)
        floors = floors[0][1]
    except:
        floors = ''
    try:
        state = re.findall(r'(Stan).*\n(.*)', parameters[0].text)
        state = state[0][1]
    except:
        state = ''
    try:
        build_yr = re.findall(r'(Rok budowy).*\n(.*)', parameters[0].text)
        build_yr = build_yr[0][1]
    except:
        build_yr = ''
    try:
        build_type = re.findall(r'(Typ zabudowy).*\n(.*)', parameters[0].text)
        build_type = build_type[0][1]
    except:
        build_type = ''
    try:
        build_material = re.findall(r'(Materiał budynku).*\n(.*)', parameters[0].text)
        build_material = build_material[0][1]
    except:
        build_material = ''
    try:
        parking = re.findall(r'(Miejsce parkingowe).*\n(.*)', parameters[0].text)
        parking = parking[0][1]
    except:
        parking = ''
    try:
        parking_spots = re.findall(r'(Liczba miejsc parkingowych).*\n(.*)', parameters[0].text)
        parking_spots = parking_spots[0][1].replace(' ','') 
    except:
        parking_spots = ''        
    try:
        kitchen = re.findall(r'(Forma kuchni).*\n(.*)', parameters[0].text)
        kitchen = kitchen[0][1]
    except:
        kitchen = ''
    try:
        windows = re.findall(r'(Okna).*\n(.*)', parameters[0].text)
        windows = windows[0][1]  
    except:
        windows = ''     
    try:
        ownership = re.findall(r'(Forma własności).*\n(.*)', parameters[0].text)
        ownership = ownership[0][1] 
    except:
        ownership = ''
    try:
        transport = re.findall(r'(Komunikacja).*\n(.*)', parameters[0].text)
        transport = transport[0][1].replace(', ','*') 
    except:
        transport = '' 
    try:
        education = re.findall(r'(Edukacja).*\n(.*)', parameters[0].text)
        education = education[0][1].replace(', ','*') 
    except:
        education = ''   
    try:
        health = re.findall(r'(Zdrowie i uroda).*\n(.*)', parameters[0].text)
        health = health[0][1].replace(', ','*') 
    except:
        health = ''         
    try:
        recreation = re.findall(r'(Rekreacja).*\n(.*)', parameters[0].text)
        recreation = recreation[0][1].replace(', ','*') 
    except:
        recreation = '' 
    try:
        other = re.findall(r'(Pozostałe).*\n(.*)', parameters[0].text)
        other = other[0][1].replace(', ','*') 
    except:
        other = ''
    try:
        other2 = re.findall(r'(Udogodnienia).*\n(.*)', parameters[0].text)
        other2 = other2[0][1].replace(', ','*') 
    except:
        other2 = ''
    try:
        additional_area = re.findall(r'(Powierzchnia dodatkowa).*\n(.*)', parameters[0].text)
        additional_area = additional_area[0][1].replace(', ','*') 
    except:
        additional_area = ''
    try:
        bathroom = re.findall(r'(wyposażenie łazienki).*\n(.*)', parameters[0].text)
        bathroom = bathroom[0][1].replace(', ','*') 
    except:
        bathroom = '' 
    try:
        installation = re.findall(r'(Stan instalacji).*\n(.*)', parameters[0].text)
        installation = installation[0][1] 
    except:
        installation = '' 
    try:
        noise = re.findall(r'(Głośność).*\n(.*)', parameters[0].text)
        noise = noise[0][1].replace(', ','*') 
    except:
        noise = ''
    try:
        heating = re.findall(r'(Ogrzewanie i energia).*\n(.*)', parameters[0].text)
        heating = heating[0][1].replace(', ','*') 
    except:
        heating = ''
    try:
        media = re.findall(r'(Media).*\n(.*)', parameters[0].text)
        media = media[0][1].replace(', ','*') 
    except:
        media = ''  
    try:
        direction = re.findall(r'(Usytuowanie względem stron świata).*\n(.*)', parameters[0].text)
        direction = direction[0][1].replace(', ','*') 
    except:
        direction = '' 
    try:
        price_info = soup.find_all('span',class_="priceInfo__value")
        price = price_info[0].text.replace(' ','').replace('\n','').replace(',','.')[:-2]
    except:
        price = ''

    # data collection
    data.append([city,
                district,
                voivodeship,
                localization_y,
                localization_x,
                market,
                area,
                rent,
                rooms,
                floor,
                floors,
                state,
                build_yr,
                build_type,
                build_material,
                parking,
                parking_spots,
                kitchen,
                windows,
                ownership,
                transport,
                education,
                health,    
                recreation,
                other,
                other2,
                additional_area,
                bathroom, 
                installation, 
                noise,
                heating,
                media,  
                direction,
                price,
                url])
    
    i += 1
    if i%300 == 0:
        print(round(i/len(auction_list)*100,2),'%')
        
#     time.sleep(0.5)
        
stop = time.time()
print((stop-start)//60,'min.')
```

## saving auction details

```python
with open('apartment_details_all.csv','w') as file:
    file.write('city,district,voivodeship,localization_y,localization_x,market,area,rent,rooms,floor,floors,state,build_yr,build_type,build_material,parking,parking_spots,kitchen,windows,ownership,transport,education,health,recreation,other,other2,additional_area,bathroom,installation,noise,heating,media,direction,price,url\n')
    for line in data:
        file.write(','.join(line) + '\n')
```
